##One-hot encoding the columns with string values.

```python
copy = eis.loc[:,str_cols]
```

```python
pd.get_dummies(copy.loc[:,'Alt1_Code'])
```

```python
copy
```

```python
copy = copy.drop(['Row_index'], axis=1)
```

```python
copy
```

```python
import pandas as pd

for i in copy.columns:
  # print(f"this is the value of copy[{i}]: ", copy.loc[:,i])
  dummies = pd.get_dummies(copy.loc[:, i])
  # print("This is the value of dummies: ", dummies)
  o_cols = list(dummies.columns)
  # print("this is the value of o_cols: ", o_cols)
  cols = list(dummies.columns)
  # print("this is the value of cols: ", cols)
  for j in range(len(cols)):
    # print("this is the value of j: ", j)
    cols[j] = cols[j] + "_" + str(i)
  dictionary = dict()
  for j in range(len(o_cols)):
    dictionary[o_cols[j]] = cols[j]
  dummies.rename(columns = dictionary, inplace=True)
  copy = pd.concat([copy, dummies],axis=1)
```

```python
len(copy.columns)
```

```python
all_columns = list(copy.columns)
```

```python
Empty_cols = []
for i in range(len(all_columns)):
  if all_columns[i][0] == 'X':
    Empty_cols.append(all_columns[i])
```

```python
Empty_cols
```

```python
copy = copy.drop(Empty_cols, axis=1)
```

```python
len(copy.columns)
```

```python
copy = copy.drop(str_cols, axis=1)
```
